```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
def lemmatize_sentence(tokens):
    lemmatizer = WordNetLemmatizer()
    lemmatized_sentence = []
    for word, tag in pos_tag(tokens):
        if tag.startswith('NN'):
            pos = 'n'
        elif tag.startswith('VB'):
            pos = 'v'
        else:
            pos = 'a'
        lemmatized_sentence.append(lemmatizer.lemmatize(word, pos))
    return lemmatized_sentence

print(lemmatize_sentence(tweet_tokens[0]))
```

```python
import re, string
```

```python
def remove_noise(tweet_tokens, stop_words = ()):

    cleaned_tokens = []

    for token, tag in pos_tag(tweet_tokens):
        token = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\(\),]|'\
                       '(?:%[0-9a-fA-F][0-9a-fA-F]))+','', token)
        token = re.sub("(@[A-Za-z0-9_]+)","", token)
        
        token = re.sub("#", "",token)
        
        token = re.sub("(.)\1{1,}", "\1\1" , token)
    
        if tag.startswith("NN"):
            pos = 'n'
        elif tag.startswith('VB'):
            pos = 'v'
        else:
            pos = 'a'

        lemmatizer = WordNetLemmatizer()
        token = lemmatizer.lemmatize(token, pos)

        if len(token) > 0 and token not in string.punctuation and token.lower() not in stop_words:
            cleaned_tokens.append(token.lower())
    return cleaned_tokens
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
print(remove_noise(tweet_tokens[0], stop_words))
```

```python
positive_tweet_tokens = twitter_samples.tokenized('positive_tweets.json')
```

```python
negative_tweet_tokens = twitter_samples.tokenized('negative_tweets.json')
```

```python
positive_cleaned_tokens_list = []
```

```python
negative_cleaned_tokens_list = []
```

```python
for tokens in positive_tweet_tokens:
    positive_cleaned_tokens_list.append(remove_noise(tokens, stop_words))
```

```python
print(positive_cleaned_tokens_list)
```

```python
for tokens in negative_tweet_tokens:
    negative_cleaned_tokens_list.append(remove_noise(tokens, stop_words))
```

```python
print(negative_cleaned_tokens_list)
```

```python
def get_all_words(cleaned_tokens_list):
    for tokens in cleaned_tokens_list:
        for token in tokens:
            yield token
```

```python
all_pos_words = get_all_words(positive_cleaned_tokens_list)
```

```python
from nltk import FreqDist
```

```python
freq_dist_pos = FreqDist(all_pos_words)
print(freq_dist_pos.most_common(50))
```

```python
all_neg_words = get_all_words(negative_cleaned_tokens_list)
```

```python
freq_dist_neg = FreqDist(all_neg_words)
print(freq_dist_neg.most_common(50))
```

```python
def get_tweets_for_model(cleaned_tokens_list):
    for tweet_tokens in cleaned_tokens_list:
        yield dict([token, True] for token in tweet_tokens)
```

```python
positive_tokens_for_model = get_tweets_for_model(positive_cleaned_tokens_list)
```

```python
negative_tokens_for_model = get_tweets_for_model(negative_cleaned_tokens_list)
```

```python
import random
```

```python
positive_dataset = [(tweet_dict, "Positive")
                     for tweet_dict in positive_tokens_for_model]
```

```python
negative_dataset = [(tweet_dict, "Negative")
                     for tweet_dict in negative_tokens_for_model]
```

```python
dataset = positive_dataset + negative_dataset
```

```python
random.shuffle(dataset)
```

```python
train_data = dataset[:8000]
test_data = dataset[8000:]
```

```python
from nltk import classify
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(train_data)
```

```python
print("Accuracy is:", 100*classify.accuracy(classifier, test_data))
```

```python
print(classifier.show_most_informative_features(100))
```
